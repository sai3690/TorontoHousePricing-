```python
print('Data shape before procedure: ', dataset.shape)

dataset.dropna(subset=['final_price'], inplace=True)
dataset.drop_duplicates(subset=['mls'], inplace=True)

print('Data shape after procedure: ', dataset.shape)

for column in ['final_price', 'list_price']:
    dataset[column] = dataset[column].apply(lambda b: b.replace(',', '').replace('$', '')).astype('int32')
```

```python
dataset.head()
```

```python
trouble_addresses = dataset[dataset.full_address == 'nan, Ontario, Canada'].index
print(trouble_addresses)

for address_index in trouble_addresses.tolist():
    address = dataset.loc[address_index, 'title']
    address_a = address.find('(')
    dataset.loc[address_index,'full_address'] = address[:(address_a-1)] + str(', Ontario, Canada')
```

```python
dataset = dataset[dataset['type']!='Link']
```

```python
dataset['bedrooms'] = dataset['bedrooms'].apply(lambda a: a.replace('beds',''))
dataset['bathrooms'] = dataset['bathrooms'].apply(lambda a: a.replace('baths',''))
dataset['parking'] = dataset['parking'].apply(lambda a: a.replace('parking',''))
dataset['parking'] = dataset['parking'].apply(lambda a: a.replace('no','0'))

dataset = dataset[dataset['sqft']!='N/A sq. ft.']

```

```python
dataset.head()
```

```python
dataset['bathrooms'] = dataset['bathrooms'].astype(int)
dataset['bedrooms'] = dataset['bedrooms'].apply(lambda a: a.replace('+',''))
dataset['bedrooms'] = dataset['bedrooms'].apply(lambda a: a.replace(' ',''))
dataset['bedrooms'] = dataset['bedrooms'].astype(int)
```

```python

new_bedroom=[]
for column in dataset['bedrooms']:
                      res = sum(map(int, str(column)))
                      new_bedroom.append(res)


dataset['new_bedroom'] = new_bedroom
dataset.head(10)
```

```python
new_sqft=[]
for column in dataset['sqft']:
                      result_1 = column.split('–', 2)[0]
                      new_sqft.append(result_1)


dataset['new_sqft'] = new_sqft
dataset.head(1)

```

```python
dataset.drop('sqft' ,axis=1,inplace= True)
dataset.drop('list_price' ,axis=1,inplace= True)
dataset.drop('bedrooms' ,axis=1,inplace= True)
dataset.head(1)
```

```python
dataset['new_sqft'] = dataset['new_sqft'].apply(lambda a: a.replace('0-499','499'))
dataset['new_sqft'].value_counts()
dataset['new_sqft'] = dataset['new_sqft'].apply(lambda a: a.replace('499 sq. ft.','499'))
```

```python

dataset['new_sqft'] = dataset['new_sqft'].astype(int)
dataset.corr()
```

```python

import geopy
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
from geopy import distance

```

```python
geocoder = RateLimiter(Nominatim(user_agent='tutorial').geocode, min_delay_seconds=1)
dataset['Location'] = dataset['full_address'].apply(geocoder)

# add latitude and longitude to dataframe
dataset['Latitude'] = dataset['Location'].apply(lambda loc: loc.latitude if loc else None)
dataset['Longitude'] = dataset['Location'].apply(lambda loc: loc.longitude if loc else None)

```
